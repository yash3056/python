Q1: Find average Marks
Write a program to input marks of three tests of a student (all integers). Then calculate and print the average of all test marks.


```python
x=int(input("marks of test 1 : "))
y=int(input("marks of test 2 : "))
z=int(input("marks of test 3 : "))
a=(x+y+z)/3
print("Average marks for 3 test is ",a)
```

Q2: Find X raised to power N
You are given two numbers ’x’(it’s a float), and ’n’(it’s a integer).

Your task is to calculate ‘x’ raised to power ‘n’, and return it.

Q3: Check Palindrome
Given a string, determine if it is a palindrome, considering only alphanumeric characters.

```python
name='naman'
namerev=name[::-1]
if (name==namerev):
    print('it is Palindrome')
else:
    print('it is not Palindrome')
```

```python
s1 = "Global Warming"
#part a
print("answer for part a",s1[-1:-5:-1])
#part b
print("answer for part b",s1[4:8:1])
#part c
print("answer for part c",s1.isalnum())
#part d
print("answer for part d",s1[0:-4:1])
#part e
print("answer for part e",s1[4::1])
#part f
print("answer for part f",s1.find("Wa"))
#part g
print("answer for part g",s1.swapcase())
#part h
print("answer for part h",s1.istitle())
#part i
print("answer for part i",s1.replace("a","*"))
```
